```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import cos, sin
```

```python
def f(x):
    return np.tan(x)
```

```python
def df(x):
    return np.cos(x)**(-2)

def d2f(x):
    return 2*(np.cos(x)**(-3))*sin(x)
```

```python
def fch(x):
    global h
    return (f(x+h)-f(x-h))/(2*h)
```

```python
h=0.0001

x=np.arange(-1.5, 1.5, h)
```

```python
y=f(x)
plt.plot(x, y)
```

```python
dy=df(x)
plt.plot(x, dy)
```

```python
ych=fch(x)
plt.plot(x, ych)
```

```python
plt.plot(x, abs(dy-ych))
```

```python
e_mash=2**-53
a=np.pi/4
e=np.linspace(a-h, a+h, 10000000)
Ml=max(d2f(e))
Mo=max(f(e))

h_opt=2*(Mo*e_mash/Ml)**0.5

print(h_opt)
```

```python
E=[]
h_mass=np.arange(h_opt/10, 10*h_opt, h_opt/1000)
for h in h_mass:
    e=np.linspace(a-h, a+h, 10000)
    Ml=max(d2f(e))
    Mo=max(f(e))
    err_met=Ml*h/2
    err_round=(Mo*2)*e_mash/h
    #err_min=2*(Mo*Ml*e_mash)**0.5
    E+=[err_met+err_round]
    #E+=[err_min]
```

```python
plt.plot(h_mass, np.array(E))
#plt.plot(h_opt, )
plt.show()
```
