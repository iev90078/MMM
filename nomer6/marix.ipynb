# Метод Гаусса

```python
def gauss(rashir):
    n=len(rashir)
    for i in range(n):
        k=rashir[i][i]
        rashir[i]=rashir[i]/k
        for j in range(i+1, n):
            rashir[j]-=rashir[i]*rashir[j][i]
        #print(rashir)
    
    for i in range(n-1, -1, -1):
        rashir[i]/=rashir[i][i]
        for j in range(i-1, -1, -1):
            rashir[j]-=rashir[i]*rashir[j][i]

        #print(rashir)
    return  [rashir[i][-1] for i in range(100)]



```

```python
ans=gauss(rashir)
print(ans)
```

```python
nevyazka=np.dot(a, ans)-f
print(nevyazka)
```

```python
norma=0
for elem in nevyazka:
    norma+=elem**2
norma=norma**1/2
print(norma)
```

# Число обусловленности

```python
def norma_matrix(a):
    max=0
    for string in a:
        s=0
        for elem in string:
            s+=abs(elem)
        if s>max:
            max=s
    return max

normaa=norma_matrix(a)
```

```python
normaa=norma_matrix(a)
normaa1=norma_matrix(np.linalg.inv(a))
print(normaa*normaa1, normaa1, normaa)
```
